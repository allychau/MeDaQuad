```python
Total=poverty_16[['Postal Code','Poverty Percent, All Ages']]
Total.head()
```

```python
x_axis = np.arange(len(poverty_16))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis, poverty_16["Poverty Percent, All Ages"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, poverty_16["Postal Code"], rotation="vertical")
```

```python
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(poverty_16["Poverty Percent, All Ages"])+10)
```

```python
plt.title("Poverty % by State")
plt.xlabel("State")
plt.ylabel("% Poverty")
```

```python
plt.show()
```
